This notebook is neither very ordered nor easily reproducable. It was used in large hurry close to the deadline to create the numbers for the final runs.
To this end, the `at.medunigraz.imi.bst.trec.SigirPubmedRecallExperimenterOptimizedBoosting2017` and `at.medunigraz.imi.bst.trec.SigirPubmedRecallExperimenterOptimizedBoosting2018` classes were used to run the desired experiments. The results and stats were just written into the `results/` and `stats/` directories, without distinguishing between the years. Thus, the tables of this notebook were created from run-specific contents of the `stats` directory. Older tables wouldn't be the same since the contents of the `stats/` directory changed. That being said, the stats won't be saved anyway so the experiments would have to be re-run first.

```python
import pandas as pd
import os
```

```python
def prepareStats(path):
    """
    Reads a single stats CSV file, excludes the 'all' row and converts the topic numbers to ints.
    Then sets the Topic columns as the new index.
    Returns a DataFrame indexed by the non-'all' topics.
    """
    df = pd.read_csv(path)
    df = df.set_index("Topic")
    return df
```

# 2017 phrases

```python
runstatsmap = {}
statspath = "../../stats/"
for stat in sorted(list(filter(lambda f: f.endswith(".csv"), os.listdir(statspath)))):
    df = prepareStats(statspath+stat)
    run = stat.replace("OFFICIAL_", "").replace(".csv", "")
    runstatsmap[run] = df
allstats = pd.concat(runstatsmap)
allstats.query("Topic == 'all'")
```
