```python
def mapindex(tup):
    m = {"all_must":"and", "all_should":"or","baseline_must": "and", "baseline_should": "or", "pmclass2018.keyword": "LogReg", "pmclass2018lstmgru.keyword": "DL", "baseline": "baseline", "":"",  "pmclass2017.keyword": "LogReg", "pmclass2017lstmgru.keyword": "DL",}
    tmp = tup[0]
    cls = tup[1]
    return (m[tmp], m[cls])
table = df.query("template.str.contains('baseline') and not classifier.str.contains('"+str(year)+"')").groupby(["template", "classifier"]).mean()
table = table.drop(["relgs", "relfound", "ndcg", "P_10", "year"], axis=1)
table = table.reset_index()
table.loc["baseline"] = blline
table["template"].iat[-1] = "baseline"
table["classifier"].iat[-1] = ""
#print(table.to_latex(index=False))
table = table.set_index(["template", "classifier"])
table.index = table.index.map(mapindex)
table.index.names = ["op", "model"]
latex = table.round(1).to_latex()
addMidruleToLatex(latex)
```
